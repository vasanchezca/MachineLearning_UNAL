### Funciones:

```python
def scale(factor,ds):
    upscale_factor = factor
    # Caculate new height and width using upscale_factor
    new_width = ds.rio.width * upscale_factor
    new_height = ds.rio.height * upscale_factor
    #upsample raster
    up_sampled = ds.rio.reproject(ds.rio.crs, shape=(int(new_height), int(new_width)), resampling=Resampling.bilinear)
    return up_sampled
```

```python
from rasterio.warp import reproject, Resampling, calculate_default_transform
import rasterio

# Esta funci칩n sirve para llevar todos los datos a la misma resoluci칩n especial a partir de un archivo base.

def reproj_match(infile, match, outfile):
    """Reproject a file to match the shape and projection of existing raster. 
    
    Parameters
    ----------
    infile : (string) path to input file to reproject
    match : (string) path to raster with desired shape and projection 
    outfile : (string) path to output file tif
    """
    # open input
    with rasterio.open(infile) as src:
        src_transform = src.transform
        
        # open input to match
        with rasterio.open(match) as match:
            dst_crs = match.crs
            
            # calculate the output transform matrix
            dst_transform, dst_width, dst_height = calculate_default_transform(
                src.crs,     # input CRS
                dst_crs,     # output CRS
                match.width,   # input width
                match.height,  # input height 
                *match.bounds,  # unpacks input outer boundaries (left, bottom, right, top)
            )

        # set properties for output
        dst_kwargs = src.meta.copy()
        dst_kwargs.update({"crs": dst_crs,
                           "transform": dst_transform,
                           "width": dst_width,
                           "height": dst_height,
                           "nodata": 0})
        print("Coregistered to shape:", dst_height,dst_width,'\n Affine',dst_transform)
        # open output
        with rasterio.open(outfile, "w", **dst_kwargs) as dst:
            # iterate through bands and write using reproject function
            for i in range(1, src.count + 1):
                reproject(
                    source=rasterio.band(src, i),
                    destination=rasterio.band(dst, i),
                    src_transform=src.transform,
                    src_crs=src.crs,
                    dst_transform=dst_transform,
                    dst_crs=dst_crs,
                    resampling=Resampling.nearest)
```

```python
# Se debe ingresar manualmente el nombre y ruta de los archivos para hacer la conversi칩n

LS   = "ERA5_evapo/evapototal.tif"
base = "ERA5_wind/wind_5.tif" # base

# co-register LS to match base raster
reproj_match(infile = LS, 
             match= base,
             outfile = "ERA5_evapo/evapototal_new.tif")
```

```python
ds_air_temp = xr.open_dataset('Temp_air/air_temp_con_std_este_es.nc')
ds_air_temp.rio.write_crs('epsg:4326', inplace = True)
```

```python
ds_air_temp = ds_air_temp.transpose('time', 'y', 'x')
df_clds = ds_clouds_upscaled.to_dataframe()
```

```python
slices = df_clds.loc[:,:,"2003-01-01":"2014-12-01"].drop(['spatial_ref'],axis=1)
```

```python
slices['air_temp'] = np.nan
slices['climatologia'] = np.nan
slices['anomalias'] = np.nan
slices['desviacion'] = np.nan
```

```python
slices
```

```python
temp_max_mensual = pd.read_csv(f"temperatura_aire_2003_2014.csv", ',', index_col='Fecha')
temp_max_mensual = temp_max_mensual.drop(['CodigoEstacion','Altitud'], axis=1)
temp_max_mensual['Latitud'] = np.round(temp_max_mensual['Latitud'].values,2)
temp_max_mensual['Longitud'] = np.round(temp_max_mensual['Longitud'].values,2)
temp_max_mensual
```

```python
lats = np.unique(temp_max_mensual['Latitud'])
lons = np.unique(temp_max_mensual['Longitud'])
lonss = np.round(slices.index.levels[0],2)
latss = np.round(slices.index.levels[1],2)
```

```python
for x in lats:
    if x in latss:
        x2 = x
    elif np.round(x+0.01,2) in latss:
        x2 = np.round(x+0.01,2)
    elif np.round(x-0.01,2) in latss:
        x2 = np.round(x-0.01,2)
    elif np.round(x+0.02,2) in latss:
        x2 = np.round(x+0.02,2)
    elif np.round(x-0.02,2) in latss:
        x2 = np.round(x-0.02,2)
    elif np.round(x+0.03,2) in latss:
        x2 = np.round(x+0.03,2)
    elif np.round(x-0.03,2) in latss:
        x2 = np.round(x-0.03,2)
    elif np.round(x+0.04,2) in latss:
        x2 = np.round(x+0.04,2)
    elif np.round(x-0.04,2) in latss:
        x2 = np.round(x-0.04,2)
    else:
        x2 = np.nan
    for y in lons:
        if np.round(y,2) in lonss:
            y2 = np.round(y,2)
        elif np.round(y+0.01,2) in lonss:
            y2 = np.round(y+0.01,2)
        elif np.round(y-0.01,2) in lonss:
            y2 = np.round(y-0.01,2)
        elif np.round(y+0.02,2) in lonss:
            y2 = np.round(y+0.02,2)
        elif np.round(y-0.02,2) in lonss:
            y2 = np.round(y-0.02,2)
        elif np.round(y+0.03,2) in lonss:
            y2 = np.round(y+0.03,2)
        elif np.round(y-0.03,2) in lonss:
            y2 = np.round(y-0.03,2)
        elif np.round(y+0.04,2) in lonss:
            y2 = np.round(y+0.04,2)
        elif np.round(y-0.04,2) in lonss:
            y2 = np.round(y-0.04,2)
        else:
            y2 = np.nan
        datos = temp_max_mensual[temp_max_mensual['Longitud'] == y]
        datos_val  = datos[datos['Latitud'] == x]['Valor'].values
        datos_cl   = datos[datos['Latitud'] == x]['Climatologia'].values
        datos_anom = datos[datos['Latitud'] == x]['Anomalia'].values
        datos_std  = datos[datos['Latitud'] == x]['DosSTD'].values
        # print('val', datos_val)
        # print('clm', datos_cl)
        if len(datos_val) != 0 and len(datos_val) == 144:
            if len(datos_val) == 144:
                # print(datos)
                posx = list(latss).index(x2)
                posy = list(lonss).index(y2)
                coorx = slices.index.levels[1][posx]
                coory = slices.index.levels[0][posy]
                
                slices['air_temp'].loc[coory,coorx,:] = datos_val  
                slices['climatologia'].loc[coory,coorx,:] = datos_cl  
                slices['anomalias'].loc[coory,coorx,:] = datos_anom
                slices['desviacion'].loc[coory,coorx,:] = datos_std
```

```python
sli = slices.to_xarray()
```
